```python
# Import the NetworkX library to create network graph.
import networkx as nx

# Import matplotlib's pyplot module to draw the graph.
import matplotlib.pyplot as plt
```

```python
# Create an undirected graph (that is, relationships go both ways).
G = nx.Graph()

# Create nodes. Each node represents a person in the network.
nodes = [
    "Carol", "Andre", "Fernando", "Diane",
    "Beverly", "Garth", "Ed",
    "Heather", "Ike", "Jane"
]

# This adds all nodes from the list above to the graph.
G.add_nodes_from(nodes)

# Next, I create a list of edges (connections) between two nodes (people).
edges = [
    ("Carol", "Andre"),
    ("Carol", "Fernando"),
    ("Carol", "Diane"),
    ("Andre", "Fernando"),
    ("Andre", "Diane"),
    ("Andre", "Beverly"),
    ("Fernando", "Diane"),
    ("Fernando", "Garth"),
    ("Diane", "Beverly"),
    ("Diane", "Garth"),
    ("Beverly", "Ed"),
    ("Garth", "Ed"),
    ("Garth", "Beverly"),
    ("Garth", "Heather"),
    ("Fernando", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

# Now I add all edges to the graph.
G.add_edges_from(edges)
```

```python
# Create a graph.
# I start by plotting a new window with a specific size.
# figsize is measured in inches (width, height).

plt.figure(figsize=(10, 6))

# I used a spring layout to visualize the graph.
# The advantage of using this visualization is that:
# - nodes repel each other to avoid overlap.
# I set a seed for reproducibility.
pos = nx.spring_layout(G, seed=123)  

# Next, I draw the graph, G.
nx.draw(
    G, pos,
    with_labels=True,
    node_size=2000,
    node_color="lightblue",
    font_size=9,
    edge_color="gray"
)

plt.title("Krackhardt Kite Graph")
plt.show()
```
