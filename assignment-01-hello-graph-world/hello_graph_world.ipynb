```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an undirected graph.
G = nx.Graph()

# Create nodes.
nodes = [
    "Carol", "Andre", "Fernando", "Diane",
    "Beverly", "Garth", "Ed",
    "Heather", "Ike", "Jane"
]

G.add_nodes_from(nodes)

# Create edges.
edges = [
    ("Carol", "Andre"),
    ("Carol", "Fernando"),
    ("Carol", "Diane"),
    ("Andre", "Fernando"),
    ("Andre", "Diane"),
    ("Andre", "Beverly"),
    ("Fernando", "Diane"),
    ("Fernando", "Garth"),
    ("Diane", "Beverly"),
    ("Diane", "Garth"),
    ("Beverly", "Ed"),
    ("Garth", "Ed"),
    ("Garth", "Beverly"),
    ("Garth", "Heather"),
    ("Fernando", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

G.add_edges_from(edges)
```

```python
# Create a graph.

plt.figure(figsize=(10, 6))

# layout for visualization

pos = nx.spring_layout(G, seed=42)  

nx.draw(
    G, pos,
    with_labels=True,
    node_size=2000,
    node_color="lightblue",
    font_size=9,
    edge_color="gray"
)

plt.title("Krackhardt Kite Graph")
plt.show()
```
